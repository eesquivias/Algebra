Deberas crear tu propia funci칩n para calcular el error de reconstrucci칩n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
#Genero la funci칩n y compruebo con el ejemplo que obtengo el mismo resultado. 
### TODO: Funci칩n para calcular el error de reconstrucci칩n
### TU RESPUESTA ABAJO
import numpy as np
#A = misc.face(gray=True)
A= np.array([[1, 2], [3, 4]])
A_hat = np.array([[1.01, 1.75], [2.81, 3.99]])

def sse_score(A, A_hat):
    SSE = np.sum((A - A_hat)**2)
    return SSE
sse_score(A, A_hat)
    
```

Una vez que ya tenemos la funci칩n `sse` hecha, podemos pasar a construir la funci칩n que ejecutar치 `SVM`.

```python
#Genero funci칩n y compruebo con el ejemplo
### TODO: Funci칩n para ejecutar SVM
### Tiene como entrada una matriz X
### Devuelve U, s, Vt

### Hint: S debe ser una matriz diagonal
### TU RESPUESTA ABAJO
A= np.array([[1, 2], [3, 4]])

def svm(A):
    U, s, Vt = np.linalg.svd(A, full_matrices=False)
    S= np.diag(s)
    return U,S,Vt

svm(A)


```

Como hemos visto en clase, las matrices obtenidas a partir de `SVM` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci칩n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
#Igual que en los ejercicios anteriores, creo funcion y compruebo que funciona
### TODO: Funci칩n para reconstruir la matriz original a partir de U, s, Vt
### Tiene como entrada U, s, Vt
### Devuelve X_hat
### TU RESPUESTA ABAJO
U = np.array([[-0.40455358, -0.9145143 ],[-0.9145143 ,  0.40455358]])
S = np.array([[5.4649857 , 0.],[0., 0.36596619]])
Vt = np.array([[-0.57604844, -0.81741556],[ 0.81741556, -0.57604844]])

def reconstruction(U, S, Vt):
    A_hat_reconstructed = U.dot(S.dot(Vt))
    return A_hat_reconstructed

reconstruction(U, S, Vt)
        
    
```

Calcula el error de reconstrucci칩n usando la funci칩n `sse` que has programado anteriormente.

```python
sse = sse_score(A, A_hat)
print(f"El error de reconstrucci칩n es: {sse}")
```

```python
#Tras varias pruebas opto por la formulaci칩n que dejo abajo, quitando los valores que se salen del rango, creando una variable que los englobe y luego elimin치ndola. Con el resto de opciones me costaba m치s conseguir un dataframe de resultado. 

### TODO: Funci칩n para crear un nuevo subset en base a valores.
### Como argumento recibe un dataframe y una lista de tuples
### Tuples: (column_name, min_value, max_value)
### Devuelve un dataframe que excluye las filas donde los valores, en la columna que le hemos indicado, exceden los valores
### que le hemos indicado
### No eliminar la fila si los valores son iguales al min/max valor
### TU RESPUESTA ABAJO
cutoffs = [('SalePrice', 50000, 1e10), ('GrLivArea', 0, 4000)]
def column_cutoff(df_sub, cutoffs):  
    out= df_sub[(df_sub.GrLivArea>cutoffs[1][2])| (df_sub.GrLivArea<cutoffs[1][1]) | (df_sub.SalePrice<cutoffs[0][1]) | (df_sub.SalePrice> int(cutoffs[0][2]))].index
    df_sub_cutoff= df_sub.drop(out, inplace= False)
    return df_sub_cutoff

df_sub_cutoff=column_cutoff(df_sub,cutoffs)
print(len(df_sub_cutoff))
df_sub_cutoff.head()
```

#### M칤nimos Cuadrados / Least Squares

Ahora, implementar치s la ecuaci칩n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
### TODO: Funci칩n para resolver la ecuaci칩n wLS
### Toma como argumentos dos matrices, una para X y otra para y
### Asumimos que las matrices tienen las dimensiones correctas
### Paso 1: Asegurate que n > d. 
X = df_sub_cutoff['GrLivArea'].values
Y = df_sub_cutoff['SalePrice'].values

print(X.shape)
print(Y.shape)
#print(type(X))

#para las transformaciones de datos defino una funci칩n y as칤 puedo recurrir a ella si cambian los datos. y me sirve para el punto siguiente
def transformacion_datos (X,Y):
    Y= np.array([Y])
    X= np.array([X])
    X=np.transpose(X)
    Y=np.transpose(Y)
    #print(Y)
#las observaciones son m치s que las dimensiones.
### Paso 2: Debes a침adir a la matriz X un vector columna del tama침o (n x 1)
    X= np.append(arr=np.ones([1451, 1]).astype(int), values=X,axis=1)
    #print(X)
    #print(type(X))
    #print(X.shape)
    #print(Y.shape)
### Paso 3: Usa la ecuaci칩n de arriba para obtener wLS
### TU RESPUESTA ABAJO
    input_x= X
    #type(input_x)
    target_y= Y
    return input_x,target_y

input_x,target_y= transformacion_datos(X,Y)

def least_squares_weights(input_x, target_y):
    input_x_trans= np.transpose(input_x)
    #print(input_x_trans)
    prod= np.dot(input_x_trans,input_x)
    inv_prod= np.linalg.inv(prod)
    inv_prod_trans= np.dot(inv_prod,input_x_trans)
    #print(inv_prod_trans)
    inv_prod_trans_y=np.dot(inv_prod_trans,target_y)
    return inv_prod_trans_y

weights = least_squares_weights(input_x, target_y)
print(weights)

#mi resultado es el mismo que el de sklearn!!
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matem치ticos detr치s del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresi칩n.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m치ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi칩n lineal, se basa en resolver esta ecuaci칩n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

#### Leer Datos

```python
import pandas as pd
import numpy as np

# Leer datos
data = pd.read_csv('./data/train.csv')

# Extraer dichas columnas
newData = data[['GrLivArea','SalePrice']]
print(newData.head())

# Contruir x - y
x = newData['GrLivArea']
y = newData['SalePrice']

# Standarizar los datos
x = (x - x.mean()) / x.std()
x = np.c_[np.ones(x.shape[0]), x] 

print("Shape of X: ", x.shape)
print("Shape of y:", y.shape)
```

#### Gradient Descent

```python
### TODO: Funci칩n para encontrar los valores w usando Gradient Descent
### Toma como argumentos: X, y, w, n_iterations, eta
### Completa la funci칩n a침adiendo la loss funci칩n y la updating rule
### TU RESPUESTA ABAJO
#Xt= np.transpose(x)
#print(Xt)

eta = 0.01 
iterations = 2000

#y = training_y
#x = training_x
np.random.seed(123)
w = np.random.rand(2)
#print(y.shape)
past_loss = []
past_w = [w]
n = y.size

xt= np.transpose(x)

def gradient_descent(x, y, w, iterations, eta):
    for i in range(iterations):
        prediction = np.dot(x, w)
        error = prediction - y
        loss= (1/2/n)*(np.dot(np.transpose(error),error))
        past_loss.append(loss)
        GradRss = np.dot(xt, error)
        w = w - (1/n)*eta*GradRss
        past_w.append(w)
    return past_w, past_loss

weights, loss = gradient_descent(x, y, w, iterations, eta)
print(weights[-1])

print(loss[-1])

#print(GradRss) 
    

    
    
```

Una vez construida nuestra funci칩n para el Gradient Descent podemos usarla para encontrar los valores optimos de $w$. **Prueba a modificar el learning rate para ver la convergencia del Gradient Descent.**

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Funci칩n para convertir un conjunto a vector
    
    Argumentos:
        tokens_web1 -- Contenido tokenizado de p치gina web 1
        tokens_web_2 -- Contenido tokenizado de p치gina web 2
        set_web1 -- Conjunto de palabras de la p치gina web 1
        set_web2 -- Conjunto de palabras de la p치gina web 2
        
    Ejemplo:
        tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
        tokens_web_2 = ["hola", "me", "llamo"]
        set_web1 = {"hola", "que", "tal"}
        set_web2 = {"hola", "me", "llamo"}
        union_ls = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
        
    Requerimientos:
        Depende de la funci칩n `union()`
    """
    
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    web1_array = []
    web2_array = [] 

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return web1_array, web2_array

tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
tokens_web_2 = ["hola", "me", "llamo"]
set_web1 = {"hola", "que", "tal", "soy", "Alex"}
set_web2 = {"hola", "me", "llamo"}
web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
```
